#Question 1: Distance Matrix Calculation

```python
import pandas as pd
import numpy as np

def calculate_distance_matrix(file_path):
    df = pd.read_csv(file_path)

    unique_ids = sorted(set(df['id_start']).union(df['id_end']))
    distance_matrix = pd.DataFrame(float('inf'), index=unique_ids, columns=unique_ids)

    np.fill_diagonal(distance_matrix.values, 0)

    for index, row in df.iterrows():
        id_start = row['id_start']
        id_end = row['id_end']
        distance = row['distance']

        distance_matrix.loc[id_start, id_end] = distance
        distance_matrix.loc[id_end, id_start] = distance

    for k in unique_ids:
        for i in unique_ids:
            for j in unique_ids:
                if distance_matrix.loc[i, j] > distance_matrix.loc[i, k] + distance_matrix.loc[k, j]:
                    distance_matrix.loc[i, j] = distance_matrix.loc[i, k] + distance_matrix.loc[k, j]

    pd.set_option("display.precision", 1)
    return distance_matrix


```

```python
file_path = 'dataset-3.csv'
result_matrix = calculate_distance_matrix(file_path)

result_matrix
```

#Question 2: Unroll Distance Matrix

```python
def unroll_distance_matrix(distance_matrix):
    unrolled_df = pd.DataFrame(columns=['id_start', 'id_end', 'distance'])

    for i in distance_matrix.index:
        for j in distance_matrix.columns:
            if i != j:
                row = {'id_start': i, 'id_end': j, 'distance': distance_matrix.loc[i, j]}
                unrolled_df = pd.concat([unrolled_df, pd.DataFrame([row])], ignore_index=True)

    return unrolled_df
```

```python
unrolled_df = unroll_distance_matrix(result_matrix)

unrolled_df
```

#Question 3: Finding IDs within Percentage Threshold

```python
import pandas as pd

def find_ids_within_ten_percentage_threshold(df, reference_id):
    reference_data = df[df['id_start'] == reference_id]

    reference_avg_distance = reference_data['distance'].mean()

    threshold_lower = reference_avg_distance - 0.1 * reference_avg_distance
    threshold_upper = reference_avg_distance + 0.1 * reference_avg_distance
    similar_ids = df[(df['id_start'] != reference_id) & (df['distance'] >= threshold_lower) & (df['distance'] <= threshold_upper)]
    similar_ids = similar_ids.sort_values(by='id_start').reset_index(drop=True)

    return similar_ids


```

```python
reference_id = 1001402
result_df = find_ids_within_ten_percentage_threshold(unrolled_df, reference_id)

result_df
```

#Question 4: Calculate Toll Rate

```python
def calculate_toll_rate(unrolled_df):
    df_with_toll_rates = unrolled_df.copy()

    rate_coefficients = {'moto': 0.80, 'car': 1.20, 'rv': 1.50, 'bus': 2.20, 'truck': 3.60}

    for vehicle_type, rate_coefficient in rate_coefficients.items():
        df_with_toll_rates[vehicle_type] = round(df_with_toll_rates['distance'] * rate_coefficient, 2)

    df_with_toll_rates = df_with_toll_rates.drop(columns=['distance'])

    return df_with_toll_rates


```

```python
result_df_with_toll_rates = calculate_toll_rate(unrolled_df)

result_df_with_toll_rates
```

#Question 5: Calculate Time-Based Toll Rates

```python
import pandas as pd
import datetime

def calculate_time_based_toll_rates(input_df):
    df = input_df.copy()

    weekday_ranges = [
        (datetime.time(0, 0, 0), datetime.time(10, 0, 0), 0.8),
        (datetime.time(10, 0, 0), datetime.time(18, 0, 0), 1.2),
        (datetime.time(18, 0, 0), datetime.time(23, 59, 59), 0.8)
    ]
    weekend_ranges = [
        (datetime.time(0, 0, 0), datetime.time(23, 59, 59), 0.7)
    ]

    id_start_list, id_end_list, distance_list = [], [], []
    start_day_list, start_time_list, end_day_list, end_time_list = [], [], [], []
    moto_list, car_list, rv_list, bus_list, truck_list = [], [], [], [], []

    for index, row in df.iterrows():
        id_start, id_end, distance = row['id_start'], row['id_end'], row['distance']

        for day in range(7):
            day_of_week = datetime.datetime.strptime(str(day), "%w").strftime('%A')

            if day_of_week in ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']:
                time_ranges = weekday_ranges
            else:
                time_ranges = weekend_ranges
            for start_time, end_time, discount_factor in time_ranges:
                id_start_list.append(id_start)
                id_end_list.append(id_end)
                distance_list.append(distance)
                start_day_list.append(day_of_week)
                start_time_list.append(start_time)
                end_day_list.append(day_of_week)
                end_time_list.append(end_time)

                moto_list.append(round(distance * discount_factor * 0.67, 2))
                car_list.append(round(distance * discount_factor * 1, 2))
                rv_list.append(round(distance * discount_factor * 1.25, 2))
                bus_list.append(round(distance * discount_factor * 1.83, 2))
                truck_list.append(round(distance * discount_factor * 3, 2))

    result_df = pd.DataFrame({
        'id_start': id_start_list,
        'id_end': id_end_list,
        'distance': distance_list,
        'start_day': start_day_list,
        'start_time': start_time_list,
        'end_day': end_day_list,
        'end_time': end_time_list,
        'moto': moto_list,
        'car': car_list,
        'rv': rv_list,
        'bus': bus_list,
        'truck': truck_list
    })

    return result_df

```

```python
result_df_with_time_based_rates = calculate_time_based_toll_rates(unrolled_df)

result_df_with_time_based_rates
```
