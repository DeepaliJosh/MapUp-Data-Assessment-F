#Question 2: Car Type Count Calculation

```python
def get_type_count(df):

  df['car_type'] = df['car'].apply(lambda x: 'low' if x <= 15 else ('medium' if x <= 25 else 'high'))
  type_counts = df['car_type'].value_counts().to_dict()

  return type_counts
```

```python
car_type_counts = get_type_count(df)

# Display the resulting dictionary of car type counts
print(car_type_counts)
```

#Question 3: Bus Count Index Retrieval

```python
def get_bus_indexes(df):

    bus_mean = df['bus'].mean()

    bus_indexes = df[df['bus'] > 2 * bus_mean].index.tolist()

    return bus_indexes

result_bus_indexes = get_bus_indexes(df)

# Display the resulting list of bus indexes
print(result_bus_indexes)
```

#Question 6: Time Check

```python
import pandas as pd
from itertools import product
from tqdm.auto import tqdm

tqdm.pandas()

file_path = 'dataset-2.csv'
df = pd.read_csv(file_path, encoding='UTF-8-SIG')


def check_time_completeness(dataframe):

  all_days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
  all_hours = [f'{hour:02}:00:00' for hour in range(24)]


  unique_ids = dataframe['id'].unique()
  unique_id_2s = dataframe['id_2'].unique()
  multi_index = pd.MultiIndex.from_product([unique_ids, unique_id_2s], names=['id', 'id_2'])

  completeness_series = pd.Series(False, index=multi_index)

  grouped = dataframe.groupby(['id', 'id_2'])
  for (id_val, id_2_val), group in tqdm(grouped, desc='Checking completeness'):

      days_covered = group['startDay'].isin(all_days).all() and group['endDay'].isin(all_days).all()

      hours_covered = group['startTime'].isin(all_hours).all() and group['endTime'].isin(all_hours).all()

      completeness_series.loc[(id_val, id_2_val)] = days_covered and hours_covered

  return completeness_series


completeness_check = check_time_completeness(df)

print(completeness_check.head())

count_values = completeness_check.value_counts()
count_values
```
